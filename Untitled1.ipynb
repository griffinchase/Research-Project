```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


def sdot (s,t,params):
    
    #state variables
    S, I, R = s
    
    #parameter names
    rzero, Ti = params
    
    #model ODEs
    
    dS = -(I/Ti)*rzero*S
    dI = (I/Ti)*((rzero*S)-1)
    dR = I/Ti
    
    ds= [dS, dI, dR]
    
    return ds
    
    
#initial conditions

S0= 0.9999
I0= 0.0001
R0= 0

#parameter values

rzero= 15.0
Ti= 8.0

s0= [S0,I0,R0]

params= [rzero, Ti]

t_max = 30
t_obs = np.linspace(0, t_max, 1000)

s_obs = odeint(sdot,s0,t_obs, args=(params,))

S_obs = s_obs[:,0]
I_obs = s_obs[:,1]
R_obs = s_obs[:,2]

#plot results (seperate)

fig1 = plt.figure()
axS=fig1.add_subplot(3,1,1)
axI=fig1.add_subplot(3,1,2)
axR=fig1.add_subplot(3,1,3)

axS.plot(t_obs, S_obs, 'b-')
axI.plot(t_obs, I_obs, 'r-')
axR.plot(t_obs, R_obs, 'g-')

axS.set_ylabel("susceptible")
axI.set_ylabel("infectious")
axR.set_ylabel("recovered")
axR.set_xlabel("time (years)")
```

```python

def sdot(s, t, rzero, Ti):
    S, I, R = s
    dS = -(I/Ti)*rzero*S
    dI = (I/Ti)*((rzero*S)-1)
    dR = I/Ti
    return [dS, dI, dR]

def simulate_sir(rzero, Ti, s0, t_obs):
    return odeint(sdot, s0, t_obs, args=(rzero, Ti))

def abc_rejection(data, num_samples, threshold):
    accepted_samples = []
    for _ in range(num_samples):
        rzero_prior = np.random.uniform(14, 16)  
        Ti_prior = np.random.uniform(7, 9)  
        simulated_data = simulate_sir(rzero_prior, Ti_prior, s0, t_obs)
        distance = np.linalg.norm(data - simulated_data)
        if distance < threshold:
            accepted_samples.append([rzero_prior, Ti_prior])
    return np.array(accepted_samples)

# Observed data
S_observed = s_obs[:, 0]
I_observed = s_obs[:, 1]
R_observed = s_obs[:, 2]

data_observed = np.stack([S_observed, I_observed, R_observed], axis=-1)
accepted_params = abc_rejection(data_observed, num_samples=10000, threshold=0.1)  

# Plotting the accepted parameters
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].hist(accepted_params[:, 0], bins=50, color='blue', edgecolor='black')
ax[0].set_title('Histogram of Accepted rzero Values')
ax[0].set_xlabel('rzero')
ax[0].set_ylabel('Frequency')

ax[1].hist(accepted_params[:, 1], bins=50, color='red', edgecolor='black')
ax[1].set_title('Histogram of Accepted Ti Values')
ax[1].set_xlabel('Ti')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Analyzing accepted parameters
mean_rzero = np.mean(accepted_params[:, 0])
mean_Ti = np.mean(accepted_params[:, 1])

print(f"Mean rzero: {mean_rzero}")
print(f"Mean Ti: {mean_Ti}")
```
